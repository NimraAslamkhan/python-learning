# hangman game

```python
# hangman game
import random

# -----------------------------
# WORD LIST
# -----------------------------
words = ["apple", "tiger", "banana", "car", "python", "school", "friend", "garden", "camera"]

# -----------------------------
# PICK A RANDOM WORD
# -----------------------------
word = random.choice(words)
word_letters = list(word)
blanks = ["_"] * len(word)

attempts = 6
guessed_letters = []

print("üéâ Welcome to Hangman Game!")
print("Guess the word letter by letter.")
print("You have", attempts, "attempts.\n")

# -----------------------------
# MAIN GAME LOOP
# -----------------------------
while attempts > 0 and "_" in blanks:

    print("Word:", " ".join(blanks))
    print("Attempts left:", attempts)
    print("Guessed letters:", guessed_letters)

    guess = input("\nEnter a letter: ").lower()

    # Check input validity
    if len(guess) != 1 or not guess.isalpha():
        print("‚ùå Please enter ONE alphabet letter only.\n")
        continue

    if guess in guessed_letters:
        print("‚ö†Ô∏è You already guessed that letter!\n")
        continue

    guessed_letters.append(guess)

    # -----------------------------
    # CHECK IF GUESS IS CORRECT
    # -----------------------------
    if guess in word_letters:
        print("‚úî Correct guess!\n")

        # Reveal the guessed letter in the blanks
        for i in range(len(word)):
            if word[i] == guess:
                blanks[i] = guess
    else:
        attempts -= 1
        print("‚ùå Wrong guess! Attempts left:", attempts, "\n")

# -----------------------------
# END OF GAME
# -----------------------------
if "_" not in blanks:
    print("üéâ CONGRATULATIONS! You guessed the word:", word)
else:
    print("üíÄ GAME OVER! The correct word was:", word)
```

# caeser cipher game by using functions

```python
# Simple Caesar Cipher Game

# -----------------------------
# Ask user for choice
# -----------------------------
print("Welcome to Caesar Cipher Game!")
print("1. Encrypt a message")
print("2. Decrypt a message")

choice = input("Enter 1 or 2: ")

# -----------------------------
# Ask user for message and shift
# -----------------------------
message = input("Enter your message: ")
shift = int(input("Enter shift number (e.g., 3): "))

# -----------------------------
# Function to encrypt
# -----------------------------
def encrypt(text, shift):
    result = ""
    for char in text:
        if char.isalpha():  # Check if character is a letter
            # Shift lowercase letters
            if char.islower():
                result += chr((ord(char) - 97 + shift) % 26 + 97)
            # Shift uppercase letters
            else:
                result += chr((ord(char) - 65 + shift) % 26 + 65)
        else:
            result += char  # Keep numbers, symbols, spaces as is
    return result

# -----------------------------
# Function to encrypt
# -----------------------------
def decrypt(text, shift):
    result = ""
    for char in text:
        if char.isalpha():
            if char.islower():
                result += chr((ord(char) - 97 - shift) % 26 + 97)
            else:
                result += chr((ord(char) - 65 - shift) % 26 + 65)
        else:
            result += char
    return result

# -----------------------------
# Main logic
# -----------------------------
if choice == "1":
    print("Encrypted message:", encrypt(message, shift))
elif choice == "2":
    print("Decrypted message:", decrypt(message, shift))
else:
    print("Invalid choice! Please enter 1 or 2.")
```

# Black Jeck Game

```python
import random

# Deal one random card
def deal_card():
    cards = [11, 2, 3, 4, 5, 6, 7, 8, 9, 10, 10, 10, 10]  # 11 = Ace
    return random.choice(cards)

# Calculate score
def calculate_score(cards):
    # Check for Blackjack (Ace + 10)
    if sum(cards) == 21 and len(cards) == 2:
        return 0

    # Convert Ace (11) to 1 if score > 21
    if 11 in cards and sum(cards) > 21:
        cards.remove(11)
        cards.append(1)

    return sum(cards)

# Compare scores
def compare(user_score, computer_score):
    if user_score == computer_score:
        return "Draw üòê"
    elif computer_score == 0:
        return "You lose! Opponent has Blackjack üò≠"
    elif user_score == 0:
        return "You win with a Blackjack! üéâ"
    elif user_score > 21:
        return "You went over 21. You lose üò¢"
    elif computer_score > 21:
        return "Opponent went over 21. You win üéâ"
    elif user_score > computer_score:
        return "You win üéâ"
    else:
        return "You lose üò¢"

def play_game():
    user_cards = []
    computer_cards = []
    is_game_over = False

    # Starting two cards each
    for _ in range(2):
        user_cards.append(deal_card())
        computer_cards.append(deal_card())

    # Game loop
    while not is_game_over:
        user_score = calculate_score(user_cards)
        computer_score = calculate_score(computer_cards)

        print(f"Your cards: {user_cards}, current score: {user_score}")
        print(f"Computer's first card: {computer_cards[0]}")

        # If Blackjack or bust
        if user_score == 0 or computer_score == 0 or user_score > 21:
            is_game_over = True
        else:
            should_continue = input("Type 'y' to get another card, 'n' to pass: ")

            if should_continue == "y":
                user_cards.append(deal_card())
            else:
                is_game_over = True

    # Computer draws until score >= 17 (and not Blackjack)
    while computer_score != 0 and computer_score < 17:
        computer_cards.append(deal_card())
        computer_score = calculate_score(computer_cards)

    print(f"\nYour final hand: {user_cards}, final score: {user_score}")
    print(f"Computer's final hand: {computer_cards}, final score: {computer_score}")
    print(compare(user_score, computer_score))


# Start the game
play_game()
```

```python
# -------------------------------
#  Class: Course
# -------------------------------
class Course:
    def __init__(self, course_name, credit_hours):
        self.course_name = course_name
        self.credit_hours = credit_hours


# -------------------------------
#  Class: Student
# -------------------------------
class Student:
    def __init__(self, name, student_id):
        self.name = name
        self.student_id = student_id
        self.__marks = {}   # Encapsulated (Private)

    def enroll(self, course):
        self.__marks[course.course_name] = None

    def add_marks(self, course_name, marks):
        if course_name in self.__marks:
            self.__marks[course_name] = marks

    def calculate_grade(self, marks):
        if marks >= 90:
            return "A"
        elif marks >= 80:
            return "B"
        elif marks >= 70:
            return "C"
        elif marks >= 60:
            return "D"
        else:
            return "F"

    def get_report(self):
        print(f"\nStudent: {self.name} (ID: {self.student_id})")
        print("Course Report:")
        for course, marks in self.__marks.items():
            if marks is None:
                print(f" - {course}: Not Graded yet")
            else:
                grade = self.calculate_grade(marks)
                print(f" - {course}: {marks} ‚Üí Grade: {grade}")


# -------------------------------
#  Inheritance Example
# -------------------------------
class ScholarshipStudent(Student):
    def __init__(self, name, student_id, scholarship_percent):
        super().__init__(name, student_id)
        self.scholarship_percent = scholarship_percent

    # Polymorphism (different fee structure)
    def get_fee(self, base_fee):
        discounted = base_fee - (base_fee * self.scholarship_percent / 100)
        return discounted


# -------------------------------
#  School Class (Manager)
# -------------------------------
class School:
    def __init__(self):
        self.students = []
        self.courses = []

    def add_student(self, student):
        self.students.append(student)

    def add_course(self, course):
        self.courses.append(course)

    def show_all_students(self):
        print("\nAll Students:")
        for stu in self.students:
            print(f"- {stu.name} ({stu.student_id})")


# -------------------------------
# Usage Example
# -------------------------------
school = School()

# Create courses
math = Course("Mathematics", 3)
python = Course("Python Programming", 4)

school.add_course(math)
school.add_course(python)

# Create students
s1 = Student("Nimra", "ST001")
s2 = ScholarshipStudent("Ayesha", "ST002", scholarship_percent=50)

school.add_student(s1)
school.add_student(s2)

# Enroll students
s1.enroll(math)
s1.enroll(python)

s2.enroll(math)

# Add marks
s1.add_marks("Mathematics", 92)
s1.add_marks("Python Programming", 88)

s2.add_marks("Mathematics", 75)

# Show Reports
s1.get_report()
s2.get_report()

# Scholarship fee example
print("\nFee Calculation for Scholarship Student:")
print("Original Fee: 40,000")
print("Discounted Fee:", s2.get_fee(40000))
```

List comprehension is a concise and powerful way to create lists in Python. It allows you to generate a new list from an existing iterable (like a list, range, or string) in one readable line.

# Project: Countdown Timer Machine (Pomodoro Style)
üîÅ Timer Cycle

‚è± 25 minutes work

‚òï 10 minutes break

‚è± 25 minutes work again

(Keeps repeating)

## 1Ô∏è‚É£ Project Logic (English ‚Äì Exam Ready)

Create a Tkinter window.

Display a timer label on the screen.

Start a 25-minute countdown when user clicks Start.

When 25 minutes finish:

Automatically start 10-minute break.

After break ends:

Start 25-minute work session again.

Timer updates every second.

User can reset the timer anytime.

2Ô∏è‚É£ Important Concepts Used

‚úî Tkinter window
‚úî Label
‚úî Button
‚úî after() method (for countdown)
‚úî State control (Work / Break)

```python
import tkinter as tk

# ---------------- CONSTANTS ----------------
WORK_TIME = 25 * 60     # 25 minutes
BREAK_TIME = 10 * 60    # 10 minutes

timer_running = False
is_work_time = True
remaining_time = WORK_TIME

# ---------------- FUNCTIONS ----------------
def start_timer():
    global timer_running
    if not timer_running:
        timer_running = True
        countdown()

def reset_timer():
    global timer_running, is_work_time, remaining_time
    timer_running = False
    is_work_time = True
    remaining_time = WORK_TIME
    title_label.config(text="Work Time", fg="green")
    timer_label.config(text="25:00")

def countdown():
    global remaining_time, is_work_time, timer_running

    if not timer_running:
        return

    minutes = remaining_time // 60
    seconds = remaining_time % 60
    timer_label.config(text=f"{minutes:02d}:{seconds:02d}")

    if remaining_time > 0:
        remaining_time -= 1
        window.after(1000, countdown)
    else:
        # Switch between work and break
        if is_work_time:
            is_work_time = False
            remaining_time = BREAK_TIME
            title_label.config(text="Break Time", fg="red")
        else:
            is_work_time = True
            remaining_time = WORK_TIME
            title_label.config(text="Work Time", fg="green")

        window.after(1000, countdown)

# ---------------- UI SETUP ----------------
window = tk.Tk()
window.title("Pomodoro Countdown Timer")
window.geometry("350x250")

title_label = tk.Label(
    window,
    text="Work Time",
    font=("Arial", 18, "bold"),
    fg="green"
)
title_label.pack(pady=10)

timer_label = tk.Label(
    window,
    text="25:00",
    font=("Arial", 40, "bold")
)
timer_label.pack(pady=10)

start_btn = tk.Button(
    window,
    text="Start",
    font=("Arial", 12),
    command=start_timer
)
start_btn.pack(pady=5)

reset_btn = tk.Button(
    window,
    text="Reset",
    font=("Arial", 12),
    command=reset_timer
)
reset_btn.pack(pady=5)

window.mainloop()
```

```python
import pandas as pd
import smtplib
from email.mime.text import MIMEText
from datetime import datetime

# ---------------- CONFIG ----------------
EMAIL_ADDRESS = "your_email@gmail.com"
EMAIL_PASSWORD = "your_password_or_app_password"

# ---------------- READ DATA ----------------
df = pd.read_csv("birthdays.csv")  # name,email,birthday

# ---------------- TODAY'S DATE ----------------
today = datetime.now()
today_str = today.strftime("%m-%d")  # '01-23'

# ---------------- CHECK AND SEND ----------------
for index, row in df.iterrows():
    # Birthday in MM-DD format
    bday = datetime.strptime(row['birthday'], "%Y-%m-%d").strftime("%m-%d")

    if bday == today_str:
        name = row['name']
        email = row['email']

        # Message
        subject = "Happy Birthday!"
        body = f"Dear {name},\n\nWishing you a very Happy Birthday üéâüéÇ\nHave a wonderful year ahead!\n\nBest wishes"

        # Email formatting
        msg = MIMEText(body)
        msg['Subject'] = subject
        msg['From'] = EMAIL_ADDRESS
        msg['To'] = email

        # Send email
        try:
            with smtplib.SMTP("smtp.gmail.com", 587) as server:
                server.starttls()
                server.login(EMAIL_ADDRESS, EMAIL_PASSWORD)
                server.send_message(msg)
            print(f"Birthday email sent to {name} ({email})")
        except Exception as e:
            print(f"Failed to send email to {name}: {e}")
```
